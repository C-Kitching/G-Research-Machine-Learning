# An introduction to supervised learning and Decision Trees

### What this session will cover...

- An intro to supervised learning
- An overview of the decision tree
- Evaluating deterministic classifiers

### What is Machine Learning?

- The field of Machine Learning seeks to answer the question "How can we build systems that **improve with data**, and what are the **fundamental laws that govern learning processes**?"
- **Machine learning** is at the intersection of **computer science** and **statistics**. Goal is to build programs, as in computer science, using data, as in statistics.

### What do we mean by **machine** and **learning**?
- A program or model
- Is trained and improves with data, instead of being hard-coded

#### Mode choice - rule-based or data driven?




### Machine learning tasks
- Machine learning tasks have at its center predicting properties for novel data
    - **Predicting** house prices
    - **Classifying** document topics
    - Grouping **similar** users in social networks
    - **Detecting** credit fraud
    - **Generating** novel music

Tasks should be quantitative properties that can be derived from the data.

### Supervised Learning
- Data contains **input** $X$ and **outcome** $y$
- Data generating process is unkown:  $y=f(X)$
- Develop a model that approximates this process and maps inputs to predictions $\hat{y} = f'(X)$
- **Regression** - continuous $y$:
    - e.g. number of rooms, postcode, m<sup class="footnote">2</sup>  ⇨  house price
- **Classification** - discrete $y$:
    - e.g. glucose concentration, blood pressure, age  ⇨  has diabetes

### Modeling steps: supervised example

- Example of modeling steps for a supervised **prediction** task:
    - Define quantitative target: e.g. *mean squared error, cross-entropy loss*
    - Choose model: e.g. *neural network, logistic regression*
    - Train model: e.g. *gradient descent*
    - Evaluate model: i.e. test predictions on novel data

a large part of these steps is automated with libraries like sklearn, and the main job in modeling is choosing a sensible model and its settings

### What to achieve from data with supervised learning? Generalisation!
- In supervised learning we want to use the available data to:
    - (1) fit predictive model (e.g. estimate the weight vector)
    - (2) quantify the model's performance
- How to do this? Usually we do this empirically. That is, we split the data in two parts:
    - Data for (1): **training set**
    - Data for (2): **test set**

### Generalization to new data
    
- It is relatively easy to perform well on the training data, like **curve fitting** with many parameters
- A fundamental concept in learning is that the model should perform well for the **unseen data** in the test set 
- Models have different levels of **complexity**, and should be chosen according to data 



### Decision trees

* Decision trees are composed of binary if-else rules similar to decision flowcharts.
* Rather than being hand-engineered, the rules are learned from data.
* The learning algorithm sets rules that separates **categories** most, e.g. 50g weight threshold for apples vs. cherries.



### Decision trees
* A **hierarchical** structure
* Each **node** corresponds to a **binary rule**
* Only **one attribute** is considered per node
* The **end** of a decision path is called a **leaf**
* **Follow** the decision path to make **predictions** for new data
* Paths don't interact, **lacking compositional** computations

### Growing the tree

How can we fit a decision tree classifier? We grow from each node, recursively, choosing the "best separation" for that node.



### Recursive binary splitting

Simple **divide-and-conquer** recursive algorithm, choosing **greedily** the best separation for each new node, until a stopping condition is met.



Key questions:
- How do we quantify the **"best separation"**?
- When do we **stop** creating nodes?

### Separation quality

* At each node, the decision tree should separate data samples that reach that node.



### When to stop splitting? Pruning criteria

If we keep splitting forever, we get end with completely sorted leaf nodes **(overfitting)**:
* Unique (*almost!*) decision path for every observation in sample
* Does not generalize well to new data

To prevent overfitting we control the complexity of the tree with pruning criteria:
   * **maximum depth** of any branch
   * **minimum number of samples** needed in a new node/leaf
   * **minimum Gini gain** for a node split
    
How to choose the parameters:
* Evaluate performance on the **validation set**

### Pros and cons of decision trees

**Pros:**
* Easy to **interpret**
* No need to **scale** data - no comparisons between features
* Can handle **both** **numerical** and **categorical** data
* **Compact**
* Making predictions is **fast** - scales with the tree depth
 
**Cons:**
* Can **overfit** easily (high variance)
* **Not robust**: small variations in data can result in very different results
* **Biased** if the dataset is **imbalanced**
* Not capable of outputting well calibrated class **probabilities**


### Evaluating a model

* A trained model's output: $\hat{y}=f'(X)$ (*target predictions*)
* Tries to approximate the *ground truth*: $y = f(X)$
* Some metric that compares predictions to the ground truth $G(y,\hat{y})$
    * Must be evaluated on data the model has never seen before

### Classification

* Binary classification (True/False):
    - $y \in \{0,1\}$
    
* Multinomial classification ($i=1,...,J$ classes):
    - $y \in \{0,1\}^J$
    - $y_i \in \{0,1\}$

* Deterministic or 0-1 Classification:
    - $\hat{y} \in \{0,1\}$


### Accuracy - how many did we get right?

* $\text{accuracy} = \frac{\# \text{correct}}{\# \text{data}}$

* More formally, $\frac{\sum_n y_n\hat{y}_n+(1-y_n)(1-\hat{y}_n)}{N}$

* Note - accuracy can be problematic for many reasons, more on that later!



### Accuracy and overfitting

* **Train accuracy**: accuracy achieved on data model is fitted to
* **Test accuracy**: accuracy achieved on data never seen before by model
* If $\text{train accuracy}\gg\text{test accuracy}$, overfitting has occured
* May need to simplify model for regularisation, e.g. restrict depth, etc
